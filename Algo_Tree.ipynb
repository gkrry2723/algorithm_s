**1. 이진검색트리(BST) 만들기**
- 특징
   - 각 노드의 왼쪽 child들 : 노드의 값보다 작음
   - 각 노드의 오른쪽 child들  : 노드의 값보다 큼
   - BST의 연산 시간 복잡도: O(h) --> 여기서 문제: 최악의 경우 O(n)이 됨.
      
      그래서 사실 최악의 경우에는 별로긴하나 그런 경우는 거의 없고 average는 O(logN)임 

```
class Node:
    def __init__(self,data,parent):
        self.data = data
        self.left = self.right = None
        self.parent = parent

class BST:
    def __init__(self,val=None):
        self.root = Node(val,None)
    
    def getRoot(self):
        return self.root

    def inorder_traverse(self,root):
        res = []
        if root != None:
            res = self.inorder_traverse(root.left)
            res.append(root.data)
            res = res + self.inorder_traverse(root.right)
        return res
      
    def insert(self, val):
        if self.root == None:
            self.root = Node(val, None)
            return
        
        current = self.root
        while True:
            if current.data > val:
                if current.left:
                    current = current.left
                else:
                    current.left = Node(val, current)
                    break
            else:
                if current.right:
                    current = current.right
                else:
                    current.right =  Node(val, current)
                    break

    
    def search_r(self,root,key):
        if root == None or key == root.data:
            return root
        elif root.data > key:
            return self.search_r(root.left,key)
        else:
            return self.search_r(root.right,key)
    
    # def search_i(self,root,key):
    #     while root != None and key != root.data:
    #     if root.data > key:
    #         root = root.left
    #     else:
    #         root = root.right
    #     return root
    
    def minimum(self, root):
        while root.left != None:
            root = root.left
        return root.data
    
    def maximum(self, root):
        while root.right !=None:
            root = root.right
        return root.data
    
    # 이걸 할라면 tree에서 부모노드도 저장하고 있어야함.
    # 입력 값보다 크면서 가장 작은 값. 그 다음값!  
    def successor(self, root, key):
        node = self.search_r(root,key)
        if node.right != None:
            suc = self.minimum(node.right)
            return suc
        
        else:
            parent = node.parent
            while parent != None and node == parent.right:
                node = parent
                parent = node.parent
        return parent.data
    
    def predecessor(self, root, key):
        node = self.search_r(root, key)
        if node.left != None:
            suc = self.maximum(node.left)
            return suc
        else:
            parent = node.parent
            while parent != None and node == parent.left:
                node = parent
                parent = node.parent
            return parent.data
    
    def delete(self, root, key):
        node = self.search_r(root, key)
        #자식이 없는 경우
        parent = node.parent
        if node.left == None and node.right == None:
            if parent == None:
                self.root = None
            elif parent.left == node:
                parent.left = None
            else:
                parent.right = None
        #자식노드가 1개인 경우
        elif node.left == None and node.right != None:
            if parent == None:
                self.root = node.right
            elif parent.left == node:
                parent.left = node.right
            else:
                parent.right = node.right
        elif node.left != None and node.right == None:
            if parent == None:
                self.root = node.left
            elif parent.left == node:
                parent.left = node.left
            else:
                parent.right = node.left
        #자식노드가 2개인 경우
        else:
            suc = self.successor(root, key)
            self.delete(root,suc)
            node.data = suc



            


     



  
# tree = BST(15)
# tree.insert(6)
# tree.insert(18)
# tree.insert(3)
# tree.insert(7)
# tree.insert(17)
# tree.insert(20)
# tree.insert(2)
# tree.insert(4)
# tree.insert(13)
# tree.insert(9)
# print(tree.inorder_traverse(tree.root))
# node = tree.minimum(tree.root)
# print(node)
# node = tree.maximum(tree.root)
# print(node)
# print(tree.successor(tree.root,4))
# print(tree.predecessor(tree.root, 15))
# tree.delete(tree.root,4)
# print(tree.inorder_traverse(tree.root))
tree= BST(15)
tree.insert(16)
print(tree.inorder_traverse(tree.root))
tree.delete(tree.root,15)
print(tree.inorder_traverse(tree.root))
print(tree.root.data)
```
