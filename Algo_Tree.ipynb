**1. 이진검색트리(BST) 만들기**
- 특징
   - 각 노드의 왼쪽 child들 : 노드의 값보다 작음
   - 각 노드의 오른쪽 child들  : 노드의 값보다 큼
   - BST의 연산 시간 복잡도: O(h) --> 여기서 문제: 최악의 경우 O(n)이 됨.
      
      그래서 사실 최악의 경우에는 별로긴하나 그런 경우는 거의 없고 average는 O(logN)임 

```
class Node:
    def __init__(self,data,parent):
        self.data = data
        self.left = self.right = None
        self.parent = parent

class BST:
    def __init__(self,val=None):
        self.root = Node(val,None)
    
    def getRoot(self):
        return self.root

    def inorder_traverse(self,root):
        res = []
        if root != None:
            res = self.inorder_traverse(root.left)
            res.append(root.data)
            res = res + self.inorder_traverse(root.right)
        return res
      
    def insert(self, val):
        if self.root == None:
            self.root = Node(val, None)
            return
        
        current = self.root
        while True:
            if current.data > val:
                if current.left:
                    current = current.left
                else:
                    current.left = Node(val, current)
                    break
            else:
                if current.right:
                    current = current.right
                else:
                    current.right =  Node(val, current)
                    break

    
    def search_r(self,root,key):
        if root == None or key == root.data:
            return root
        elif root.data > key:
            return self.search_r(root.left,key)
        else:
            return self.search_r(root.right,key)
    
    # def search_i(self,root,key):
    #     while root != None and key != root.data:
    #     if root.data > key:
    #         root = root.left
    #     else:
    #         root = root.right
    #     return root
    
    def minimum(self, root):
        while root.left != None:
            root = root.left
        return root.data
    
    def maximum(self, root):
        while root.right !=None:
            root = root.right
        return root.data
    
    # 이걸 할라면 tree에서 부모노드도 저장하고 있어야함.
    # 입력 값보다 크면서 가장 작은 값. 그 다음값!  
    def successor(self, root, key):
        node = self.search_r(root,key)
        if node.right != None:
            suc = self.minimum(node.right)
            return suc
        
        else:
            parent = node.parent
            while parent != None and node == parent.right:
                node = parent
                parent = node.parent
        return parent.data
    
    def predecessor(self, root, key):
        node = self.search_r(root, key)
        if node.left != None:
            suc = self.maximum(node.left)
            return suc
        else:
            parent = node.parent
            while parent != None and node == parent.left:
                node = parent
                parent = node.parent
            return parent.data
    
    def delete(self, root, key):
        node = self.search_r(root, key)
        #자식이 없는 경우
        parent = node.parent
        if node.left == None and node.right == None:
            if parent == None:
                self.root = None
            elif parent.left == node:
                parent.left = None
            else:
                parent.right = None
        #자식노드가 1개인 경우
        elif node.left == None and node.right != None:
            if parent == None:
                self.root = node.right
            elif parent.left == node:
                parent.left = node.right
            else:
                parent.right = node.right
        elif node.left != None and node.right == None:
            if parent == None:
                self.root = node.left
            elif parent.left == node:
                parent.left = node.left
            else:
                parent.right = node.left
        #자식노드가 2개인 경우
        else:
            suc = self.successor(root, key)
            self.delete(root,suc)
            node.data = suc


  
tree = BST(15)
tree.insert(6)
tree.insert(18)
tree.insert(3)
tree.insert(7)
tree.insert(17)
tree.insert(20)
tree.insert(2)
tree.insert(4)
tree.insert(13)
tree.insert(9)
print(tree.inorder_traverse(tree.root))
# node = tree.minimum(tree.root)
# print(node)
# node = tree.maximum(tree.root)
# print(node)
# print(tree.successor(tree.root,4))
# print(tree.predecessor(tree.root, 15))
# tree.delete(tree.root,4)
# print(tree.inorder_traverse(tree.root))
# tree= BST(15)
# tree.insert(16)
# print(tree.inorder_traverse(tree.root))
# tree.delete(tree.root,15)
# print(tree.inorder_traverse(tree.root))
# print(tree.root.data)
tree.leftRotate(tree.root, 6)
print(tree.inorder_traverse(tree.root))
tree.rightRotate(tree.root, 6)
print(tree.inorder_traverse(tree.root))
```

2. 레드블랙트리 ( Red-Black Tree )

- 이진 탐색 트리의 일종
- 균형잡힌 트리 -> O(n)을 피하기 위하여 사용
- 최악의 연산 경우에도 O(logn)
- 완전히 균형잡힌 트리는 아니고 완전히 극단의 트리가 되지는 않도록 도와줌
- 조건
    1. 각 노드는 red 혹은 black 노드이다.
    2. 루트노드는 black이다.
    3. 모든 리프노드(NIL)은 black이다.
    4. red노드의 자식들은 무조건 black(red노드는 연속 x, black은 상관 x)
    5. 모든 노드에 대하여 그 노드로부터 NIL노드로 이르는 모든 경로에는 동일한 개수의 black 노드가 존재한다.
- 특징
  1.  레드블랙트리의 높이는 2log(n+1) 이하이다.
  


```
# 서치 -> binary search tree와 아예 동일

class Node:
    def __init__(self,data,parent):
        self.data = data
        self.left = self.right = None
        self.parent = parent

class BST:
    def __init__(self,val=None):
        self.root = Node(val,None)
        self.color = {}
        if val != None:
          self.color[val] = "black"
    
    def getRoot(self):
        return self.root

    def inorder_traverse(self,root):
        res = []
        if root != None:
            res = self.inorder_traverse(root.left)
            res.append(root.data)
            res = res + self.inorder_traverse(root.right)
        return res


    def search_r(self,root,key):
        if root == None or key == root.data:
            return root
        elif root.data > key:
            return self.search_r(root.left,key)
        else:
            return self.search_r(root.right,key)

    def minimum(self, root):
        while root.left != None:
            root = root.left
        return root.data
    
    def maximum(self, root):
        while root.right !=None:
            root = root.right
        return root.data
    
    # 이걸 할라면 tree에서 부모노드도 저장하고 있어야함.
    # 입력 값보다 크면서 가장 작은 값. 그 다음값!  
    def successor(self, root, key):
        node = self.search_r(root,key)
        if node.right != None:
            suc = self.minimum(node.right)
            return suc
        
        else:
            parent = node.parent
            while parent != None and node == parent.right:
                node = parent
                parent = node.parent
        return parent.data
    

    def leftRotate(self, root, key):
      x = self.search_r(root,key)
      y = x.right
      if y == None:
        print("바꾸고자 하는 노드의 오른쪽 자식이 없습니다.")
        return
      else:
        x.right = y.left
        if x.right != None:
          x.right.parent = x
        y.parent = x.parent
        if x.parent == None:
          self.root = y
        elif x == x.parent.left:
          x.parent.left = y
        else:
          x.parent.right = y
        y.left = x
        x.parent = y
    
    def rightRotate(self, root, key):
      x = self.search_r(root, key)
      y = x.left
      if y == None:
        print("바꾸고자 하는 노드의 왼쪽 자식이 없습니다.")
        return
      else:
        x.left = y.right
        if x.left != None:
          x.left.parent = x
        y.parent = x.parent
        if x.parent == None:
          self.root = y
        elif x.parent.left == x:
          y.parent.left = y
        else:
          y.parent.right = y
        y.right = x
        x.parent = y
    
    def RBinsert(self, val):
        if self.root == None:
            self.root = Node(val, None)
            self.color[val] = "black"
            return
        current = self.root
        while True:
            if current.data > val:
                if current.left:
                    current = current.left
                else:
                    current.left = Node(val, current)
                    break
            else:
                if current.right:
                    current = current.right
                else:
                    current.right =  Node(val, current)
                    break
        #RB tree를 위해 추가적으로 해주는 코드
        self.color[val] = "red"
        self.RBInsertFixup(val)
      
    def RBInsertFixup(self,val):
      node = self.search_r(self.root,val)
      parent = node.parent
      while parent and self.color[parent.data] == 'red':
        # case 1,2,3
        if parent.parent.left == parent:
          #case 1
          uncle = parent.parent.right
          if uncle and self.color[uncle.data] == 'red':
            self.color[parent.data] = 'black'
            self.color[uncle.data] = 'black'
            self.color[parent.parent.data] = 'red'
            continue
          else:
            #case 2
            if node == parent.right:
              pval = parent.data
              self.leftRotate(self.root, parent.data)
              node = self.search_r(self.root,pval)
              parent = node.parent
            self.rightRotate(self.root,parent.parent.data)
            self.color[parent.data] = "black"
            self.color[parent.right.data] = "red"
            break

        # case 4,5,6
        else:
          uncle = parent.parent.left
          if uncle and self.color[uncle.data] == 'red':
            self.color[parent.data] = 'black'
            self.color[uncle.data] = 'black'
            self.color[parent.parent.data] = 'red'
            continue
          else:
            if node == parent.left:
              pval = parent.data
              self.rightRotate(self.root, parent.data)
              node = self.search_r(self.root,pval)
              parent = node.parent
            self.leftRotate(self.root, parent.parent.data)
            self.color[parent.data] = "black"
            self.color[parent.leftdata] = "red"
            break
      self.color[self.root.data] = "black"

    def RBdelete(self, root, key):
        node = self.search_r(root, key)
        #자식이 없는 경우
        parent = node.parent
        if node.left == None and node.right == None:
            x = None
            if parent == None:
                self.root = None
            elif parent.left == node:
                parent.left = None
            else:
                parent.right = None
        #자식노드가 1개인 경우
        elif node.left == None and node.right != None:
            x = node.right.data
            if parent == None:
                self.root = node.right
            elif parent.left == node:
                parent.left = node.right
            else:
                parent.right = node.right
        elif node.left != None and node.right == None:
            x = node.left.data
            if parent == None:
                self.root = node.left
            elif parent.left == node:
                parent.left = node.left
            else:
                parent.right = node.left
        #자식노드가 2개인 경우
        else:
            suc = self.successor(root, key)
            x = self.search_r(root,suc)
            self.RBdelete(root,suc)
            node.data = suc
        

        if self.color[key] == "black" and x!= None:
            del self.color[key]        
            self.RBDeleteFixup(x.data)
    
    def RBDeleteFixup(self,key):
      node = self.search_r(self.root, key)
      while node != self.root and self.color[node.data] == 'black':
        parent = node.parent
        extraBlack = key
        #case 1,2,3,4
        if parent.left == node:
          w = parent.right
          #case 1
          if self.color[w.data] == 'red':
            self.leftRotate(self.root,parent.data)
            self.color[w.data] = 'black'
            self.color[parent.data] = 'red'
            w = node.parent.right
          
          #case 2,3,4
          else:
            #case 2
            if self.color[w.right.data] == 'black' and self.color[w.left.data] =='black':
              extraBlack = parent.data
              self.color[w.data] = 'red'
              if self.color[extraBlack] == 'red':
                self.color[extraBlack] = 'black'
                break
              else:
                node = parent
                continue
            #case3,4
            else:
              #case 3
              if self.color[w.left.data] == 'red' and self.color[w.right.data] == 'black':
                self.rightRotate(self.root,w.data)
                self.color[w.data] = 'red'
                self.color[w.parent.data] = 'black'
                w = w.parent
              #case 4
              self.leftRotate(self.root,parent.data)
              self.color[w.data] = self.color[parent.data]
              self.color[parent.data] = 'black'
              extraBlack = None
              break

        #case 5,6,7,8
        else:
          w = parent.left
          #case 1
          if self.color[w.data] == 'red':
            self.rightRotate(self.root,parent.data)
            self.color[w.data] = 'black'
            self.color[parent.data] = 'red'
            w = node.parent.left
          
          #case 2,3,4
          else:
            #case 2
            if self.color[w.left.data] == 'black' and self.color[w.right.data] =='black':
              extraBlack = parent.data
              self.color[w.data] = 'red'
              if self.color[extraBlack] == 'red':
                self.color[extraBlack] = 'black'
                break
              else:
                node = parent
                continue
            #case3,4
            else:
              #case 3
              if self.color[w.right.data] == 'red' and self.color[w.left.data] == 'black':
                self.leftRotate(self.root,w.data)
                self.color[w.data] = 'red'
                self.color[w.parent.data] = 'black'
                w = w.parent
              #case 4
              self.rightRotate(self.root,parent.data)
              self.color[w.data] = self.color[parent.data]
              self.color[parent.data] = 'black'
              extraBlack = None
              break
      self.color[node.data] = 'black'
          


tree = BST(15)
tree.RBinsert(14)
tree.RBinsert(13)
tree.RBinsert(12)
tree.RBinsert(11)
print(tree.color)
root = tree.getRoot()
print(root.data)
tree.RBdelete(tree.root,12)
print(tree.inorder_traverse(tree.root))
print(tree.color)
```
